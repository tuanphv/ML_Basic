###**3. Dữ liệu trùng lặp**

```python
duplicated_rows = df[df.duplicated(keep=False)]
print(duplicated_rows)
df=df.drop_duplicates()
```

**Kết quả**

```python
print( df.duplicated().sum())
print(df.shape)
```

Dữ liệu tưởng chừng hoàn hảo ban đầu hóa ra có tổng cộng lên đến 460 mẫu trùng lặp. Sau khi xử lý các mẫu trùng lặp này bằng phương thức drop_duplicates thì dữ liệu chỉ còn lại 1359 mẫu và không còn hai mẫu nào trùng nhau nữa.
